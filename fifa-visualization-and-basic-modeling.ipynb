# <p style="background-color:purple; font-family:newtimeroman; font-size:180%; text-align:center; border-radius: 15px 50px;">  Basic Modelling üïπÔ∏è</p>


```python
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python

for col in train:
    if(train[col].dtype=='object'):
        train[col]=le.fit_transform(train[col].astype(str))
        
for col in test:
    if(test[col].dtype=='object'):
        test[col]=le.fit_transform(test[col].astype(str))

for col in train:
    if(train[col].isnull().sum()!=0):
        train[col]=train[col].fillna(train[col].dropna().median())

for col in test:
    if(test[col].isnull().sum()!=0):
        test[col]=test[col].fillna(test[col].dropna().median())
```

```python
# Split the dataframe into data and labels
df_x = train.drop('value', axis=1) # data
df_y = train.value # labels
```

```python
imputer = KNNImputer(n_neighbors=2, weights="uniform")
train = imputer.fit_transform(train)
test = imputer.fit_transform(test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=4)
```
